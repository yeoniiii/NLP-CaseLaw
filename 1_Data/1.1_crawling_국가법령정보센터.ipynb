### (2) 내용 csv - "사기"

```python
case_list = pd.read_csv('./case_list.csv')
case_list.head(3)
```

```python
case_list = case_list[~case_list.사건명.isnull()]
case_list.head(3)
```

```python
case_list[case_list.사건명.str.contains('사기')].사건명
```

```python
case_list_fraud = case_list[case_list.사건명.str.contains('사기')]
case_list_fraud = case_list_fraud[~case_list_fraud.사건명.str.contains('군사기밀|수사기관')]
print(case_list_fraud.shape)
case_list_fraud.head(1)
```

```python
contents = ['판시사항', '판결요지', '참조조문', '참조판례', '판례내용']

def remove_tag(content):
    cleaned_text = re.sub('<.*?>', '', content)
    return cleaned_text
```

```python
for content in contents:
    case_list_fraud[content] = ''

    # os.makedirs('./판례/{}'.format(content), exist_ok=True)
```

```python
case_list_fraud.reset_index(inplace=True)
```

```python
case_list_fraud.columns
```

```python
for i in trange(len(case_list_fraud)):
    url = "https://www.law.go.kr"
    link = case_list_fraud.loc[i]['판례상세링크'].replace('HTML', 'XML')
    url += link
    response = urlopen(url).read()
    xtree = ET.fromstring(response)

    for content in contents:
        text = xtree.find(content).text
        # 내용이 존재하지 않는 경우 None 타입이 반환되기 때문에 이를 처리해줌
        if text is None:
            text = '내용없음'
        else:
            text = remove_tag(text)

        case_list_fraud[content].loc[i] = text

        # file = './판례/' + content + '/' + xtree.find('판례정보일련번호').text + '.txt'
        # with open(file, 'w') as c:
        #     c.write(text)
        #     c.close()
```

```python
case_list_fraud.to_csv('./case_fraud.csv', index=False)
```
